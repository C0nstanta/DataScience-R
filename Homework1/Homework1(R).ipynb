**1. Напишіть програму, яка буде визначати, чи є номер телефону "крутим".**

"Крутий" номер - це випадки, коли:
    
- у номері підряд йдуть 4 і більше однакових цифри;
- у номері підряд йдуть 4 і більше цифри у порядку спадання чи зростання;
- сума перших 5 цифр номеру дорівнює сумі 5 останніх.

Приклади "крутих" номерів:

1. 0981177770 (випадок 1)
2. 0951234567 (випадок 2)
3. 0670171123 (випадок 3)

Програма бере на вхід строку з номером із 10 знаків. На вихід друкує або "Крутий", або "Звичайний"

```
phone_number <- as.character(readline(prompt = "Input your phone number:\n"))
sprintf("phone_number: %s", phone_number)

num_list <- as.numeric(strsplit(as.character(phone_number), "")[[1]])
print(num_list)
case_one <- function(number){
    ptrn <- "\\d*(\\d)\\1{3,9}\\d*"
    res <- grep(ptrn, number)
    
    if (!is.null(res) & length(res) > 0 ) {
        return(T)
    } else {
        return(F)
    }
}


case_two <- function(num_list){
    count_bigger <- 0
    count_less <- 0

    for (i in 2:length(num_list)){

        if (num_list[i-1] < num_list[i]){
            
            count_bigger <- count_bigger + 1
            count_less <- 0
            if (count_bigger >= 3){
                return(T)
            }
        }
    
        if (num_list[i-1] > num_list[i]){
            
            count_bigger <- 0
            count_less <- count_less +1
            if(count_less >= 3){
                return(T)
            }
        }
    }
    return(F)
}


case_three <- function(num_list){
    i = length(num_list)
    if (sum(num_list[(i-4):i]) == sum(num_list[1:5])){
        return(T)
    } else {
        return(F)
    }    
}


one <- case_one(phone_number)
two <- case_two(num_list)
three <- case_three(num_list)

if ( one == T | two == T | three == T){
    print("You have a cool phone number!")
} else {
    print("You have usual phone number!")
}

# print(one)
# print(two)
# print(three)
```
