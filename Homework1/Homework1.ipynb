**1. Напишіть програму, яка буде визначати, чи є номер телефону "крутим".**

"Крутий" номер - це випадки, коли:
    
- у номері підряд йдуть 4 і більше однакових цифри;
- у номері підряд йдуть 4 і більше цифри у порядку спадання чи зростання;
- сума перших 5 цифр номеру дорівнює сумі 5 останніх.

Приклади "крутих" номерів:

1. 0981177770 (випадок 1)
2. 0951234567 (випадок 2)
3. 0670171123 (випадок 3)

Програма бере на вхід строку з номером із 10 знаків. На вихід друкує або "Крутий", або "Звичайний"

```
import re
import numpy as np

```

```
def case_1(phone_number):
    ptrn = "\d*(\d)\\1{3,9}\d*"
    try:
        res = re.match(ptrn,phone_number)
        if res:
            # print('Your phone number is cool')
            return True, 'Case#1'
        else:
            # print("You have an usual phone number")    
            return False, 'Case#1'

    except ValueError as val:
        print(val)
        return False, 'Case#1'


def case_2(np_phone_number):
    count_bigger = 0
    count_less = 0

    for i in range(1, len(np_phone_number)):
        
        if np_phone_number[i] > np_phone_number[i-1]:
            count_less = 0
            count_bigger += 1
            if count_bigger >= 3:
                # print('This is second case, and your number is realy cool :)')
                return True, 'Case#2'

        elif np_phone_number[i] < np_phone_number[i-1]:
            count_bigger = 0
            count_less += 1
            if count_less >= 3:
                # print('This is second case, and your number is realy cool :)')
                return True, 'Case#2'
    return False, 'Case#2'


def case_3(np_phone_number):
    if sum(np_phone_number[:4]) == sum(np_phone_number[-4:]):
        # print('This is third case, and your number is cool')
        return True, 'Case#3'
    return False, 'Case#3'


phone_str = input('Input your phone number plz:\n')
if phone_str.isdigit():
    np_phone_str = np.array(list(phone_str), dtype=int)

    res1, message1 = case_1(phone_str)
    res2, message2 = case_2(np_phone_str)
    res3, message3 = case_3(np_phone_str)
    if res1 or res2 or res3:
        print(f'res1:{res1} res2:{res2} res3:{res3}')
        print('You have cool phone number')
    else:
        print('You have usual phone number')
else:
    print('Wrong phone number')
```

**2. Напишіть програму, яка буде видавати з масиву локацій координати найближчої точки до центру.**

Перший ввід - координати широт.
Другий ввід - координати довгот.

Кожне нове значення розділяйте пробілом.

Вивід - tuple з координатами довготи і широти, які найближчі до центру локацій.

Центр локацій - це середнє арифметичне введених координат довгот і широт.

Для простоти відстань міряємо в евклідовій метриці.

Всі розрахунки потрібно зробити без підключення додаткових модулів.

```
import numpy as np
```

```

latitude = input('Input latitude data plz. (Separator - space)\n')
longitude = input('Input longtitude data plz. (Separator - space)\n')

lat = np.array(list(latitude.split(" ")), dtype=float)
long = np.array(list(longitude.split(" ")), dtype=float)

point_coord = tuple(zip(lat, long))


ave_lat = sum(lat)/len(lat)
ave_long = sum(long)/len(long)


def compute_dist():
    x_coord = len(point_coord)
    dist = np.zeros((x_coord, 3), np.float)
    min_dist = np.sqrt(np.square((point_coord[0][0] - ave_long)) + 
                          np.square((point_coord[0][1] - ave_lat)))
    dist_x = point_coord[0][0]
    dist_y = point_coord[0][1]
    for i in range(1, len(point_coord)):

        dist[i][2] = np.sqrt(np.square((point_coord[i][0] - ave_long)) + 
                          np.square((point_coord[i][1] - ave_lat)))
        if dist[i][2] < min_dist:
            min_dist = dist[i][2]
            dist_x = point_coord[i][0]
            dist_y = point_coord[i][1]

        min_data = (min_dist, dist_x, dist_y)
    return min_data
 
min_data = compute_dist()
print(f'our point(dist, x, y):{min_data}')
```
