# Опис змінних

**PetID** - Унікальне hash ID тварини.

**AdoptionSpeed** - Швидкість адаптації тварини. 0 -  в той же день, коли забрали, 1 - в перший тиждень, 2 - в перший місяць, 3 - в перші три місяці, 4 - не адаптувалася за три місяці.

**Type** - Тип тварини (1 = собака, 2 = кіт).

**Name** - Кличка тварини.

**Age** - Вік тварини в місяцях.

**Breed1** - Основна порода тварини.

**Breed2** - Друга порода тварини.

**Gender** - Гендер тварини (1 = чоловік, 2 = жінка, 3 = змішано (якщо продається відразу декілька тварин)).

**Color1** - Колір 1 тварини.

**Color2** - Колір 2 тварини.

**Color3** - Колір 3 тварини.

**MaturitySize** - Розмір, коли виросте (1 = маленький, 2 = середній, 3 = веилкий, 4 = дуже великий, 0 = невідомо).

**FurLength** - Довжина шерсті (1 = коротка, 2 = середня, 3 = довга, 0 = невідомо).

**Vaccinated** - Вакцинованість тварини (1 = так, 2 = ні, 3 = невідомо).

**Dewormed** - Деглистизованість тварини (1 = так, 2 = ні, 3 = невідомо).

**Sterilized** - Стерилізованість тварини (1 = так, 2 = ні, 3 = невідомо).

**Health** - Стан здоров'я (1 = здорова, 2 = трохи травмована, 3 = сильно травмована, 0 = невідомо).

**Quantity** - Кількість тварин, що є в профілі.

**Fee** - Плата за приручення.

**State** - Місто.

**RescuerID** - Унікальне hash ID рятівника.

**VideoAmt** - Кількість відео тварини.

**PhotoAmt** - Кількість фотографій тварини.

**Description** - Опис тварини у профілі.

# 6. Закодуйте через one-hot encoder популярні породи тварин 
(тільки ті, представників яких у масиві більше 150). Але перед цим візьміть з довідника порід і замініть числові коди на назви тварин (таблиця breed_labels.csv). Те саме для більшої зрозумілості даних зробіть із кольорами тварин (таблиця color_labels.csv).

```python
# Спочатку перевиремо що в додаткових датасетах
pd.concat([breed_labels, df['Color1'], color_labels], axis=1).head(10).tail(10)
```

```python
# Це я спробував використати цикли) Залишаю як прикол и ужос ужасный)
df['Color_ID'] = df['Color1']
for i, value in enumerate(df['Color1'].values):
    for j, color_value in enumerate(color_labels['ColorID'].values):
        if value == color_value:
            df['Color_ID'][i] = color_labels['ColorName'][j]
            break          


df['Color_ID']
```

```python
# Через merge для кольорів 
#Color1
df2 = df.merge(color_labels[['ColorID', 'ColorName']], left_on='Color1', right_on='ColorID', how='left')
df2['Color1'] = np.where(df2['ColorID'], df2['ColorName'], 0)

df2.drop(['ColorID', 'ColorName'], axis=1, inplace=True)
#Color2
df2 = df2.merge(color_labels[['ColorID', 'ColorName']], left_on='Color2', right_on='ColorID', how='left')
df2['Color2'] = np.where(df2['ColorID'], df2['ColorName'], 0)
df2.drop(['ColorID', 'ColorName'], axis=1, inplace=True)
#Color3
df2 = df2.merge(color_labels[['ColorID', 'ColorName']], left_on='Color3', right_on='ColorID', how='left')
df2['Color3'] = np.where(df2['ColorID'], df2['ColorName'], 0)
df2.drop(['ColorID', 'ColorName'], axis=1, inplace=True)#, 'Color_ID'

# Через merge для породи тварини 
#Breed1
df2 = df2.merge(breed_labels[['BreedID', 'BreedName']], left_on='Breed1', right_on='BreedID', how='left')
df2['Breed1'] = np.where(df2['BreedID'], df2['BreedName'], 0)
df2.drop(['BreedID', 'BreedName'], axis=1, inplace=True)
#Breed2
df2 = df2.merge(breed_labels[['BreedID', 'BreedName']], left_on='Breed2', right_on='BreedID', how='left')
df2['Breed2'] = np.where(df2['BreedID'], df2['BreedName'], 0)
df2.drop(['BreedID', 'BreedName'], axis=1, inplace=True)


# Я не зрозумів чому замість того щоб ставити 0 воно якось насувало NaN
# тому я був вимушений окремо змінювати по всьому датасету Nan  на 0 !!! 

# df2[:] = np.where(df2.eq(np.nan), 0, df2)
df2.fillna(0, inplace=True)
df2
```

```python
# Використання one-hot-encoder та підрахунок для кожної окремої колонки
ohe_breed = pd.get_dummies(df2['Breed1'], prefix='breed1')
ohe_breed2 = pd.get_dummies(df2['Breed2'], prefix='breed2')

ohe_breed2_others = ohe_breed2.loc[:, ohe_breed2.loc[:, :].sum() < 151] # Все що меньше або дорівнює 150 breed2(спочатку думав злити все в одну колонку.. але не розумію як)) треба подумати)
ohe_breed_others = ohe_breed.loc[:, ohe_breed.loc[:, :].sum() < 151] # Все що меньше або дорівнює 150 breed1

ohe_breed2 = ohe_breed2.loc[:, ohe_breed2.loc[:, :].sum() > 150] # Все що більше 150 breed1
ohe_breed = ohe_breed.loc[:, ohe_breed.loc[:, :].sum() > 150] # Все що більше 150 breed1

ohe_color1 = pd.get_dummies(df2['Color1'], prefix='Color1')
ohe_color2 = pd.get_dummies(df2['Color2'], prefix='Color2')
ohe_color3 = pd.get_dummies(df2['Color3'], prefix='Color3')

ohe_color1 = ohe_color1.loc[:, ohe_color1.loc[:, :].sum() > 150] # Все що більше 150 color1
ohe_color2 = ohe_color2.loc[:, ohe_color2.loc[:, :].sum() > 150] # Все що більше 150 color2
ohe_color3 = ohe_color3.loc[:, ohe_color3.loc[:, :].sum() > 150] # Все що більше 150 color3

df2 = pd.concat([df2, ohe_breed, ohe_breed2, ohe_color1, ohe_color2, ohe_color3], axis=1)
df2.drop(['Breed1', 'Breed2', 'Color1', 'Color2', 'Color3', 'Color_ID'], axis=1, inplace=True)


```
